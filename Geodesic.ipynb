```python
import numpy as np

from einsteinpy.geodesic import Geodesic, Timelike, Nulllike
from einsteinpy.plotting import GeodesicPlotter, StaticGeodesicPlotter, InteractiveGeodesicPlotter
```

```python
position = [40., np.pi / 2, 0.]
momentum = [0., 0., 3.83405]
a = 0
steps = 500
delta = .1
```

```python
geod = Timelike(
    metric = "Kerr",
    metric_params = (a,),
    position=position,
    momentum=momentum,
    steps=3000,
    delta=1,
    return_cartesian=True
)

geod
```

```python
import matplotlib.pyplot as plt
gpl = GeodesicPlotter()
#gpl.savefig('test.png')
gpl.plot(geod)

gpl.show()
```

```python
gpl.save('test.png')
```
